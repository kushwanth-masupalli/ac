```python
# write a program to plot bar plot of student marks 
import matplotlib.pyplot as plt

students = ["kush","dhanish","santhosh","bhuma","maonj"]
marks = [29,40,25,36,56]

plt.bar(students,marks,color="green")
plt.title("student marks bar plot")
plt.xlabel("students")
plt.ylabel("marks")
plt.show()
```

```python
# write a program to plot line graph for temperature variation 

import matplotlib.pyplot as plt
dates = [1,2,3,4,5,6,7]
temp =  [30,32,33,31,29,28,32]

plt.plot(dates,temp,marker="*",color="green")
plt.title("temperature variation")
plt.xlabel("date")
plt.ylabel("temperature")
plt.show()
```

```python
# program to check if an given email is valid or not 
import re 

def validate_email(email):
    pattern = r'^[a-zA-Z0-9]+@[a-zA-Z0-9]+\.[a-zA-Z]{2,}$'
    return re.match(pattern, email)

email = "masuapksjdfs2020@dlkfj.com"
match = validate_email(email)

if match:
    print("Valid email:", match.group())
else:
    print("Invalid email")
```

```python
# program to print all the numbers in a given text 

import re

text = "I scored 95 marks in math, 88 in physics, and 100 in CS in 2024."

numbers = re.findall(r'\d+', text)

print("Numbers found:", numbers)
```

```python
# program to count vowels, digits, spaces, consonants 

text ="i am iron man 2 be honest i am not "

vowels = 0
digits = 0
consonants = 0
spaces = 0

for ch in text:
    if ch.lower() in 'aeiou':
        vowels += 1
    elif ch.isdigit():
        digits += 1
    elif ch == ' ':
        spaces += 1
    elif ch.isalpha():
        consonants += 1

print("Vowels:", vowels)
print("Digits:", digits)
print("Consonants:", consonants)
print("Spaces:", spaces)
```
